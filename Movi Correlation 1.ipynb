```python
#re integrating dat
df = pd.read_csv(r'C:\Users\CAREW\Documents\movies.csv')
df.head()
```

```python
#removing empty rows

df = df.dropna()
```

```python
df.head()
```

```python
#extracting years from released colunmn
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
#sorting by gross
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#correlation gross and budget

df['budget'] = df['budget'].filna(0).astype('int64')

df['gross'] = df['gross'].filna(0).astype('int64')

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')

plt.show()


```

```python
# plot budget vs gross using seaborn. line showing progression

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Lets start looking at correlation

df.corr() #there are diff types of correlation
```

```python
#visualizing correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Lets look at company , changing string to numbers 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized 
```

```python
#correlation of numerized data
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
